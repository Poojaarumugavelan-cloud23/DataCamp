```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import HTML, Image

# --- SAMPLE DATA ---
students = ['A', 'B', 'C', 'D', 'E']
marks_test1 = np.random.randint(40, 80, size=len(students))
marks_test2 = marks_test1 + np.random.randint(5, 20, size=len(students))
marks_test2 = np.clip(marks_test2, 0, 100)

# --- PLOT SETUP ---
fig, ax = plt.subplots(figsize=(7, 5))
bars = ax.bar(students, marks_test1)
ax.set_ylim(0, 100)
ax.set_ylabel("Marks (%)")
ax.set_title("Student Marks Progress")

# --- ANIMATION FUNCTION ---
def update(frame):
    progress = frame / 50
    current_marks = marks_test1 + progress * (marks_test2 - marks_test1)

    for bar, mark in zip(bars, current_marks):
        bar.set_height(mark)

    ax.set_title(f"Student Marks Progress ({int(progress * 100)}%)")
    return bars

# --- CREATE ANIMATION ---
ani = FuncAnimation(fig, update, frames=51, interval=100)

# --- DISPLAY IN COLAB ---
HTML(ani.to_jshtml())

# --- SAVE AS GIF ---
gif_writer = PillowWriter(fps=10)
ani.save("student_marks.gif", writer=gif_writer)

print("✅ GIF saved as student_marks.gif")

# --- SHOW GIF ---
Image(filename="student_marks.gif")
```
